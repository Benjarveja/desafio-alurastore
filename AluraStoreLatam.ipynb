### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
venta_tienda_1 = tienda['Precio'].sum()
venta_tienda_1
```

```python
venta_tienda_2 = tienda2['Precio'].sum()
venta_tienda_2
```

```python
venta_tienda_3 = tienda3['Precio'].sum()
venta_tienda_3
```

```python
venta_tienda_4 = tienda4['Precio'].sum()
venta_tienda_4
```

# 2. Ventas por categoría

```python
venta_categoria_tienda_1 = tienda.groupby('Categoría del Producto')['Precio'].sum()
venta_categoria_tienda_1
```

```python
venta_categoria_tienda_2 = tienda2.groupby('Categoría del Producto')['Precio'].sum()
venta_categoria_tienda_2
```

```python
venta_categoria_tienda_3 = tienda3.groupby('Categoría del Producto')['Precio'].sum()
venta_categoria_tienda_3
```

```python
venta_categoria_tienda_4 = tienda4.groupby('Categoría del Producto')['Precio'].sum()
venta_categoria_tienda_4
```

# 3. Calificación promedio de la tienda


```python
valoracion_tienda_1 = tienda['Calificación'].mean()
print(f"La calificación promedio de la tienda 1 es: {valoracion_tienda_1.round(2)}")
```

```python
valoracion_tienda_2 = tienda2['Calificación'].mean()
print(f"La calificación promedio de la tienda 2 es: {valoracion_tienda_2.round(2)}")
```

```python
valoracion_tienda_3 = tienda3['Calificación'].mean()
print(f"La calificación promedio de la tienda 3 es: {valoracion_tienda_3.round(2)}")
```

```python
valoracion_tienda_4 = tienda4['Calificación'].mean()
print(f"La calificación promedio de la tienda 4 es: {valoracion_tienda_4.round(2)}")
```

# 4. Productos más y menos vendidos

```python
producto_mas_vendido_tienda_1 = tienda['Producto'].value_counts().idxmax()
print(f"El producto mas vendido de la tienda 1 fue: {producto_mas_vendido_tienda_1}")

producto_menos_vendido_tienda_1 = tienda['Producto'].value_counts().idxmin()
print(f"El producto menos vendido de la tienda 1 fue: {producto_menos_vendido_tienda_1}")
```

```python
from numpy._core.fromnumeric import prod
producto_mas_vendido_tienda_2 = tienda2['Producto'].value_counts().idxmax()
print(f"El producto mas vendido de la tienda 2 fue: {producto_mas_vendido_tienda_2}")

producto_menos_vendido_tienda_2 = tienda2['Producto'].value_counts().idxmin()
print(f"El producto menos vendido de la tienda 2 fue: {producto_menos_vendido_tienda_2}")
```

```python
producto_mas_vendido_tienda_3 = tienda3['Producto'].value_counts().idxmax()
print(f"El producto mas vendido de la tienda 3 fue: {producto_mas_vendido_tienda_3}")

producto_menos_vendido_tienda_3 = tienda3['Producto'].value_counts().idxmin()
print(f"El producto menos vendido de la tienda 3 fue: {producto_menos_vendido_tienda_3}")
```

```python
producto_mas_vendido_tienda_4 = tienda4['Producto'].value_counts().idxmax()
print(f"El producto mas vendido de la tienda 4 fue: {producto_mas_vendido_tienda_4}")

producto_menos_vendido_tienda_4 = tienda4['Producto'].value_counts().idxmin()
print(f"El producto menos vendido de la tienda 4 fue: {producto_menos_vendido_tienda_4}")
```

# Graficado

```python
import matplotlib.pyplot as plt

# Datos para el gráfico
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
promedios = [valoracion_tienda_1, valoracion_tienda_2, valoracion_tienda_3, valoracion_tienda_4]

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(tiendas, promedios, color=['skyblue', 'lightgreen', 'salmon', 'orchid'])

# Añadir títulos y etiquetas
plt.title('Calificación Promedio por Tienda', fontsize=15)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Calificación Promedio', fontsize=12)
plt.ylim(0, 5)  # Rango de calificaciones de 1 a 5

# Añadir los valores sobre las barras
for i, v in enumerate(promedios):
    plt.text(i, v + 0.1, str(v.round(2)), textAlign='center', fontweight='bold')

plt.show()
```
