### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# Graficado

```python
import matplotlib.pyplot as plt
import pandas as pd

data_promedios = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificacion_Promedio': [valoracion_tienda_1, valoracion_tienda_2, valoracion_tienda_3, valoracion_tienda_4]
}
df_promedios = pd.DataFrame(data_promedios)

display(df_promedios)

plt.figure(figsize=(10, 6))
plt.bar(df_promedios['Tienda'], df_promedios['Calificacion_Promedio'], color=['skyblue', 'lightgreen', 'salmon', 'orchid'])

plt.title('Calificación Promedio por Tienda', fontsize=15)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Calificación Promedio', fontsize=12)
plt.ylim(0, 5)

plt.show()
```

```python
tienda['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'

df_total = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)
```

# 1. Comparativa de Ventas por Categoría entre Tiendas

```python
import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(15, 8))
sns.barplot(data=df_total, x='Categoría del Producto', y='Precio', hue='Tienda', estimator=sum, errorbar=None)
plt.title('Comparativa de Ventas por Categoría y Tienda', fontsize=15)
plt.xticks(rotation=45)
plt.ylabel('Ingresos Totales')
plt.legend(title='Tienda', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# 2. Comparativa de Métodos de Pago por Tienda

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df_total, x='Método de pago', hue='Tienda')
plt.title('Frecuencia de Métodos de Pago por Tienda')
plt.ylabel('Cantidad de Transacciones')
plt.show()
```

# 3. Comparativa de Costo de Envío Promedio

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df_total, x='Tienda', y='Costo de envío', estimator='mean', errorbar=None, palette='coolwarm', hue='Tienda', legend=False)
plt.title('Costo de Envío Promedio por Tienda')
plt.ylabel('Costo Promedio')
plt.show()
```
